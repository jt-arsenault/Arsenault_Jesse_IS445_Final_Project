# Visualizations for Professional Peers

## Imports

```python
import matplotlib.pyplot as plt
import pandas as pd
```

## 1. Read Dataset

```python
emissions_df = pd.read_csv('./SupplyChainGHGEmissionFactors_v1.2_NAICS_CO2e_USD2021.csv')
```

## 2. Get NAICS Codes by Sector (first digit) and Sub-Sector (first 3-4 digits)

### General

```python
naics_categories = ["Agriculture (" + str(len(agriculture)) + ")", 
                    "Energy/Construction (" + str(len(mining)) + ")", 
                    "Manufacturing (" + str(len(manuf)) + ")", 
                    "Trade (" + str(len(trade)) + ")",
                    "Business Services (" + str(len(finserv)) + ")",
                    "Social Services (" + str(len(social)) + ")",
                    "Arts (" + str(len(arts)) + ")",
                    "Other (" + str(len(other)) + ")"]

# Allows us to use loc to extract each group
sizes = [len(agriculture),len(mining),len(manuf),len(trade),len(finserv),len(social),len(arts),len(other)]

distance = 0
# 1
df_agr = emissions_df.loc[0:sizes[0]-1]
agr_mean = df_agr["Supply Chain Emission Factors with Margins"].mean()
agr_std = df_agr["Supply Chain Emission Factors with Margins"].std()
distance += sizes[0]

# 2
df_min = emissions_df.loc[distance:(distance + sizes[1]-1)]
min_mean = df_min["Supply Chain Emission Factors with Margins"].mean()
min_std = df_min["Supply Chain Emission Factors with Margins"].std()
distance += sizes[1]

# 3
df_mfc = emissions_df.loc[distance:(distance + sizes[2]-1)]
mfc_mean = df_mfc["Supply Chain Emission Factors with Margins"].mean()
mfc_std = df_mfc["Supply Chain Emission Factors with Margins"].std()
distance += sizes[2]

#4
df_tra = emissions_df.loc[distance:(distance + sizes[3]-1)]
tra_mean = df_tra["Supply Chain Emission Factors with Margins"].mean()
tra_std = df_tra["Supply Chain Emission Factors with Margins"].std()
distance += sizes[3]

# 5
df_ser = emissions_df.loc[distance:(distance + sizes[4]-1)]
ser_mean = df_ser["Supply Chain Emission Factors with Margins"].mean()
ser_std = df_ser["Supply Chain Emission Factors with Margins"].std()
distance += sizes[4]

# 6
df_soc = emissions_df.loc[distance:(distance + sizes[5]-1)]
soc_mean = df_soc["Supply Chain Emission Factors with Margins"].mean()
soc_std = df_soc["Supply Chain Emission Factors with Margins"].std()
distance += sizes[5]

# 7
df_art = emissions_df.loc[distance:(distance + sizes[6]-1)]
art_mean = df_art["Supply Chain Emission Factors with Margins"].mean()
art_std = df_art["Supply Chain Emission Factors with Margins"].std()
distance += sizes[6]

# 8
df_oth = emissions_df.loc[distance:(distance + sizes[7]-1)]
oth_mean = df_oth["Supply Chain Emission Factors with Margins"].mean()
oth_std = df_oth["Supply Chain Emission Factors with Margins"].std()

sector_series = [df_agr["Supply Chain Emission Factors with Margins"],
                 df_min["Supply Chain Emission Factors with Margins"],
                 df_mfc["Supply Chain Emission Factors with Margins"],
                 df_tra["Supply Chain Emission Factors with Margins"],
                 df_ser["Supply Chain Emission Factors with Margins"],
                 df_soc["Supply Chain Emission Factors with Margins"],
                 df_art["Supply Chain Emission Factors with Margins"],
                 df_oth["Supply Chain Emission Factors with Margins"]]
means = [agr_mean, min_mean, mfc_mean, tra_mean, ser_mean, soc_mean, art_mean, oth_mean]
std_array = [agr_std, min_std, mfc_std, tra_std, ser_std, soc_std, art_std, oth_std]
```

### Agriculture

```python
regex_1 = '^111.*$'
regex_2 = '^112.*$'
regex_3 = '^113.*$'
regex_4 = '^114.*$'
regex_5 = '^115.*$'

strings = emissions_df["2017 NAICS Code"].astype("string")

grouped_1 = strings.str.findall(regex_1)
grouped_2 = strings.str.findall(regex_2)
grouped_3 = strings.str.findall(regex_3)
grouped_4 = strings.str.findall(regex_4)
grouped_5 = strings.str.findall(regex_5)

# Crop Production (111) 
crops = []
for item in grouped_1:
    if item != []:
        crops.append(item)

# Animal Production, Aquaculture (112)
animals = []
for item in grouped_2:
    if item != []:
        animals.append(item)
        
# Logging and Forestry (113)
logging = []
for item in grouped_3:
    if item != []:
        logging.append(item)
        
# Fishing, Hunting, Trapping (114)
huntfish = []
for item in grouped_4:
    if item != []:
        huntfish.append(item)
        
# Support Activities (115)
support = []
for item in grouped_5:
    if item != []:
        support.append(item)
```

### Energy/Construction

```python
regex_211 = '^211.*$'
regex_212 = '^212.*$'
regex_213 = '^213.*$'
regex_2211 = '^2211.*$'
regex_2212 = '^2212.*$'
regex_2213 = '^2213.*$'
regex_223 = '^223.*$'

grouped_211 = strings.str.findall(regex_211)
grouped_212 = strings.str.findall(regex_212)
grouped_213 = strings.str.findall(regex_213)
grouped_2211 = strings.str.findall(regex_2211)
grouped_2212 = strings.str.findall(regex_2212)
grouped_2213 = strings.str.findall(regex_2213)
grouped_223 = strings.str.findall(regex_223)


# Oil and Gas Extraction (211) 
oil = []
for item in grouped_211:
    if item != []:
        oil.append(item)

# Other Mining (212)
other_mining = []
for item in grouped_212:
    if item != []:
        other_mining.append(item)
        
# Support Activities for Mining (213)
mining_support = []
for item in grouped_213:
    if item != []:
        mining_support.append(item)
        
# Electric Power (2211)
electric = []
for item in grouped_2211:
    if item != []:
        electric.append(item)
        
# Natural Gas (2212)
gas = []
for item in grouped_2212:
    if item != []:
        gas.append(item)
        
# Water and Sewage (2213)
water = []
for item in grouped_2213:
    if item != []:
        water.append(item)
        
# Construction (223)
construction = []
for item in grouped_223:
    if item != []:
        construction.append(item)
```

### Manufacturing

```python
regex_311 = '^311.*$'
regex_312 = '^312.*$'
regex_313 = '^313.*$'
regex_314 = '^314.*$'
regex_315 = '^315.*$'
regex_316 = '^316.*$'

strings = emissions_df["2017 NAICS Code"].astype("string")

grouped_311 = strings.str.findall(regex_311)
grouped_312 = strings.str.findall(regex_312)
grouped_313 = strings.str.findall(regex_313)
grouped_314 = strings.str.findall(regex_314)
grouped_315 = strings.str.findall(regex_315)
grouped_316 = strings.str.findall(regex_316)

# Food Manufacturing (311) 
food = []
for item in grouped_311:
    if item != []:
        food.append(item)

# Beverage and Tobacco (312)
beverage = []
for item in grouped_312:
    if item != []:
        beverage.append(item)
        
# Textile Mills (313)
textiles = []
for item in grouped_313:
    if item != []:
        textiles.append(item)
        
# Textile Product Mills (314)
textile_products = []
for item in grouped_314:
    if item != []:
        textile_products.append(item)
        
# Apparel Manufacturing (315)
apparel = []
for item in grouped_315:
    if item != []:
        apparel.append(item)
        
# Leather and Allied Product Manufacturing (316)
leather = []
for item in grouped_316:
    if item != []:
        leather.append(item)
```

## 3. Set up Data for Plotting

### Agriculture

```python
crops_distance = len(crops)
crop_ghg = emissions_df.loc[0:crops_distance-1]["Supply Chain Emission Factors with Margins"]

animals_distance = len(crops) + len(animals)
animal_ghg = emissions_df.loc[len(crops):animals_distance-1]["Supply Chain Emission Factors with Margins"]

logging_distance = len(crops) + len(animals) + len(logging)
logging_ghg = emissions_df.loc[len(crops)+len(animals):logging_distance-1]["Supply Chain Emission Factors with Margins"]

huntfish_distance = logging_distance + len(huntfish)
huntfish_ghg = emissions_df.loc[logging_distance:huntfish_distance-1]["Supply Chain Emission Factors with Margins"]

support_distance = huntfish_distance + len(support)
support_ghg = emissions_df.loc[huntfish_distance:support_distance-1]["Supply Chain Emission Factors with Margins"]
```

### Energy and Construction

```python
oil_distance = len(oil)
oil_ghg = emissions_df.loc[0:oil_distance-1]["Supply Chain Emission Factors with Margins"]

other_mining_distance = len(oil) + len(other_mining)
other_mining_ghg = emissions_df.loc[len(oil):other_mining_distance-1]["Supply Chain Emission Factors with Margins"]

mining_support_distance = len(oil) + len(other_mining) + len(mining_support)
mining_support_ghg = emissions_df.loc[len(oil)+len(other_mining):mining_support_distance-1]["Supply Chain Emission Factors with Margins"]

electric_distance = mining_support_distance + len(electric)
electric_ghg = emissions_df.loc[mining_support_distance:electric_distance-1]["Supply Chain Emission Factors with Margins"]

gas_distance = electric_distance + len(gas)
gas_ghg = emissions_df.loc[electric_distance:gas_distance-1]["Supply Chain Emission Factors with Margins"]

water_distance = gas_distance + len(water)
water_ghg = emissions_df.loc[gas_distance:water_distance-1]["Supply Chain Emission Factors with Margins"]

construction_distance = water_distance + len(construction)
construction_ghg = emissions_df.loc[water_distance:construction_distance-1]["Supply Chain Emission Factors with Margins"]
```

### Manufacturing

```python
food_distance = len(food)
food_ghg = emissions_df.loc[0:food_distance-1]["Supply Chain Emission Factors with Margins"]

beverage_distance = len(food) + len(beverage)
beverage_ghg = emissions_df.loc[len(food):beverage_distance-1]["Supply Chain Emission Factors with Margins"]

textiles_distance = len(food) + len(beverage) + len(textiles)
textiles_ghg = emissions_df.loc[len(food)+len(beverage):textiles_distance-1]["Supply Chain Emission Factors with Margins"]

textile_products_distance = textiles_distance + len(textile_products)
textile_products_ghg = emissions_df.loc[textiles_distance:textile_products_distance-1]["Supply Chain Emission Factors with Margins"]

apparel_distance = textile_products_distance + len(apparel)
apparel_ghg = emissions_df.loc[textile_products_distance:apparel_distance-1]["Supply Chain Emission Factors with Margins"]

leather_distance = apparel_distance + len(leather)
leather_ghg = emissions_df.loc[apparel_distance:leather_distance-1]["Supply Chain Emission Factors with Margins"]
```

# Visualizations

```python
fig, axs = plt.subplots(1, 3, figsize=(31,10))
fig.suptitle("Cross-Sector Breakdown of Supply Chain Greenhouse Gas Emissions",size=30)

# Agriculture
axs[0].set_title("Agriculture Supply Chain", size=24)
axs[0].set_xlabel("NAICS Code", size=16, labelpad=15)
axs[0].set_ylabel("Emission Factor with Margins (kg CO2e/purchaser price, 2021 USD)", size=16, labelpad=15)

ag_labels = ['111 - Crop Production', 
             '112 - Animal Production and Aquaculture', 
             '113 - Logging and Forestry', 
             '114 - Hunting, Fishing, Trapping',
             '115 - Support Services']

axs[0].scatter(crops,crop_ghg, 85, color="green",marker='o')
axs[0].scatter(animals,animal_ghg,85,color="red",marker="x")
axs[0].scatter(logging,logging_ghg,85,color="blue",marker="s")
axs[0].scatter(huntfish,huntfish_ghg,85,color="orange",marker="d")
axs[0].scatter(support,support_ghg,85,color="purple",marker="1")
axs[0].legend(labels=ag_labels,loc='upper right')


# Energy/Utilities
axs[1].set_title("Energy/Construction Supply Chain", size=24)
axs[1].set_xlabel("NAICS Code", size=16, labelpad=15)
axs[1].set_ylabel("Emission Factor with Margins (kg CO2e/purchaser price, 2021 USD)", size=16, labelpad=15)


energy_labels = ['211 - Oil & Gas Extraction', 
                 '212 - Other Mining', 
                 '213 - Mining Support', 
                 '2211 - Electricity', 
                 '2212 - Natural Gas', 
                 '2213 - Water/Sewage',
                 '223 - Construction']

axs[1].scatter(oil,oil_ghg,85,color="green",marker='o')
axs[1].scatter(other_mining,other_mining_ghg,85,color="red",marker="x")
axs[1].scatter(mining_support,mining_support_ghg,85,color="blue",marker="s")
axs[1].scatter(electric,electric_ghg,85,color="purple",marker="1")
axs[1].scatter(gas,gas_ghg,85,color="orange",marker="d")
axs[1].scatter(water,water_ghg,85,color="lightblue",marker="P")
axs[1].scatter(construction,construction_ghg,85,color="black",marker="o")
axs[1].legend(labels=energy_labels,loc='upper center')


# Manufacturing
axs[2].set_title("Manufacturing Supply Chain", size=24)
axs[2].set_xlabel("NAICS Code", size=16, labelpad=15)
axs[2].set_ylabel("Emission Factor with Margins (kg CO2e/purchaser price, 2021 USD)", size=16, labelpad=15)


manufacturing_labels = ['311 - Food Production', 
                 '312 - Beverage and Tobacco Production', 
                 '313 - Textile Mills', 
                 '314 - Textile Product Mills', 
                 '315 - Apparel', 
                 '316 - Leather & Allied Products (Footwear)']

axs[2].scatter(food,food_ghg,75,color="green",marker='o')
axs[2].scatter(beverage,beverage_ghg,75,color="red",marker="x")
axs[2].scatter(textiles,textiles_ghg,75,color="blue",marker="s")
axs[2].scatter(textile_products,textile_products_ghg,75,color="orange",marker="d")
axs[2].scatter(apparel,apparel_ghg,75,color="purple",marker="1")
axs[2].scatter(leather,leather_ghg,75,color="black",marker="o")
axs[2].legend(labels=manufacturing_labels,loc='upper right')

plt.show()
```

```python
plt.figure(figsize=(20,10))
plt.title("Supply Chain Emissions by Sector: A Detailed Breakdown (2021)", size=24)
plt.ylabel("Sector")
plt.xlabel("GHG Emission Factor (kilgorams of CO2 equivalent divided by cost)", size=16, labelpad=15)

box_text_landfill = "Solid Waste\n Landfills"
plt.text(10.5,5.4, s=box_text_landfill,color="red", backgroundcolor="white",fontsize='x-large')

box_text_hazardous = "Hazardous Waste Processing"
plt.text(1.1,5.4, s=box_text_hazardous,color="red", backgroundcolor="white",fontsize='x-large')

box_text_cement = "Cement Manufacturing"
plt.text(3.1,3.4, s=box_text_cement,color="red", backgroundcolor="white",fontsize='x-large')

box_text_beef = "Beef & Cattle, Ranching & Farming"
plt.text(2.1,1.2, s=box_text_beef,color="red", backgroundcolor="white",fontsize='x-large')

sector_labels = [
    "Agriculture",
    "Energy/Construction",
    "Manufacturing",
    "Trade",
    "Business Services",
    "Social Services",
    "Arts",
    "Other"
]

plt.boxplot(sector_series,vert=False,labels=sector_labels)
plt.show()
```
